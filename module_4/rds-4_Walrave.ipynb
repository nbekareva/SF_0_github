В настоящее время в условиях высокой конкуренции на рынке авиаперевозок важно оперативно проводить оценку эффективности выполнения авиаперевозок, используя не только финансовые показатели, такие как <u>п р и б ы л ь , р е н т а б е л ь н о с т ь, но и производственные показатели деятельности</u>. 

Все эти данные, сформированные для целей управленческого учета, позволяют оценить себестоимость перевозки одного пассажира или одного килограмма груза в разрезе рейсов и воздушных судов, а также посчитать <u>доходность от одного перевезенного пассажира или одного килограмма груза</u>. При проведении анализа эффективности авиаперевозок объектом может служить <u>вся маршрутная сеть в целом, в разбивке по регионам, городам, по отдельным маршрутам или рейсам</u>. Анализируемый период при этом может варьироваться в зависимости от целей анализа. 

В табл. 1 представлена <u>система показателей эффективности выполнения авиаперевозок</u>.

>**Показатели, характеризующие доходы:**
>
><i><u>RASK и PRASK</u></i> - характеризуют сумму полученного дохода (совокупного и билетного соответственно) на
анализируемом маршруте от максимально возможного количества кресел на воздушном судне в рассматриваемом периоде. 
>
><i><u>Yield и Yield (passenger revenue)</u></i> - удельный доход на 1 пассажиро-километр, для оценки суммы дохода (совокупного и билетного), полученного от одного перевезенного пассажира, соответственно. В отличие от RASK Yield характеризует объем полученного дохода от выполненного
пассажирооборота.
>
>**Показатели, характеризующие расходы:**
>
><i><u>CASK по совокупным и переменным расходам</u></i>. По аналогии с показателями RASK CASK характеризует сумму расходов (совокупных или переменных), приходящихся на один выполненный рейс по анализируемому маршруту или направлению.
>
>**Операционные показатели эффективности:**
>
><i><u>Load Factor</u></i> - коэффициент коммерческой загрузки, показывает долю общего количества кресел, занятых пассажирами, и позволяет определить долю перевезенной коммерческой загрузки от максимально возможной [15]. Анализ динамики данного показателя позволяет варьировать воздушными судами в зависимости от пассажиропотока на различных маршрутах и тем самым оптимизировать себестоимость перевозки.
>
><i><u>Величина среднего тарифа</u></i> рассчитывается как сумма билетного дохода, приходящаяся на одного пассажира. Иными словами, средний тариф характеризует среднюю цену билета на авиаперевозку на рассматриваемом маршруте.
>
>Операционные показатели эффективности рассматриваются в совокупности с показателями RASK и Yield, при этом они имеют тесную взаимосвязь друг с другом.
>
><b>Оценка эффективности выполнения авиаперевозок:</b>
>
><i><u>Суммы покрытия</u></i> - позволяют определить окупаемость каждой группы расходов, то есть на какой стадии производственный процесс не приносит прибыли или какие средства остаются в распоряжении предприятия после каждого этапа.
>
>Для этого рассчитывают показатели, характеризующие финансовый результат авиапредприятия (маржинальная прибыль по переменным расходам, по условно-переменным расходам и прибыль). Оценка эффективности на основе расчета маржинального дохода по переменным издержкам позволяет оценить, как выполнение конкретного рейса покрывает переменные издержки. Если сумма покрытия на рейсе является отрицательной, то руководство авиакомпании может принять решение о его закрытии. Эффект от такого управленческого решения наступает моментально, то есть убытки значительно сокращаются.

Однако авторы статьи считают, что расчет показателей, характеризующих финансовый результат от выполненных авиаперевозок, целесообразно сравнивать и оценивать не в абсолютном, а <i><u>относительном выражении</u></i>. Расчет такого показателя приведен в табл. 2.

Итого:

Доход на предельный пассажирооборот (RASK) — отношение доходов от авиаперевозок к предельному пассажирообороту.

Затраты на предельный пассажирооборот (CASK) — отношение общих операционных затрат к предельному пассажирообороту.

Предельный пассажирооборот (ASK) — мера измерения пассажирской провозной емкости, которой располагает авиакомпания,
обозначает перемещение одного пассажирского кресла на расстояние один километр, измеряется в кресло-километрах (ккм)

>https://www.rbc.ru/newspaper/2019/08/21/5d55445a9a7947a19f520c8c

=> RASK / CASK * 100%

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 50)
import seaborn as sns
from matplotlib import pyplot as plt
import datetime
```

```python
# SELECT *
# FROM
#   (SELECT *
#    FROM dst_project.flights f
#    WHERE f.departure_airport = 'AAQ'
#      AND (date_trunc('month', f.scheduled_departure) in ('2017-01-01',
#                                                          '2017-02-01',
#                                                          '2017-12-01'))
#      AND f.status not in ('Cancelled')) a
# LEFT JOIN dst_project.ticket_flights tf ON a.flight_id = tf.flight_id
# LEFT JOIN dst_project.boarding_passes bp ON tf.flight_id = bp.flight_id
# AND tf.ticket_no = bp.ticket_no
# LEFT JOIN dst_project.tickets t ON tf.ticket_no = t.ticket_no;

df = pd.read_csv('2.csv').dropna(how='all', axis=1)

for col in ['scheduled_departure', 'scheduled_arrival', 'actual_departure', 'actual_arrival']:
    df[col] = pd.to_datetime(df[col], format='%Y-%m-%dT%H:%M:%SZ')

df['day'] = df['scheduled_departure'].dt.day
df['month'] = df['scheduled_departure'].dt.month
df['dep_date'] = df['scheduled_departure'].dt.strftime(date_format='%a %d.%m')
df['dow'] = df['scheduled_departure'].dt.strftime(date_format='%a')
df['woy'] = df['scheduled_departure'].dt.strftime(date_format='%W')

df
```

```python
df.info()
```

```python
df.groupby(by=[
    'flight_no', 'aircraft_code', 'arrival_airport', 'flight_id', 'scheduled_departure',
    'dep_date', 'dow', 'woy'
#     , 'fare_conditions'
    ]).agg(pas=('passenger_name', 'count'), revenue=('amount', 'sum'))
```

```python
fl = pd.DataFrame(df.groupby(by=[
    'flight_no', 'aircraft_code', 'arrival_airport', 'flight_id', 'scheduled_departure',
    'dep_date', 'dow', 'woy'
#     , 'fare_conditions'
                    ]).agg(pas=('passenger_name', 'count'), revenue=('amount', 'sum'))
                     ).sort_values('scheduled_departure').reset_index()
fl['max_seats'] = fl['aircraft_code'].apply(lambda x: 130 if x == '733' else 97)
fl['load_factor'] = fl['pas'] / fl['max_seats']
fl['distance'] = fl['arrival_airport'].apply(lambda x: 1230 if x == 'SVO' else 639 if x == 'EGO' else 3662)
fl['ASK'] = fl['max_seats'] * fl['distance']
# fl['yield'] = fl['revenue'] / fl['ASK']
# fl['RASK'] = fl['yield'] * fl['load_factor']
fl['RASK'] = fl['revenue'] / fl['ASK']
fl['CASK'] = fl['aircraft_code'].apply(lambda x: 14 if x == 'SU9' else 9)   # взяты наугад, в статье от 2019 7 и 4 соответственно
fl['coef'] = fl['RASK'] / fl['CASK'] * 100
fl
```

```python
sns.boxplot(data=fl, x='arrival_airport', y='coef', hue='aircraft_code')
```

```python
fl[(fl['arrival_airport'] == 'SVO') & (fl['coef'] < 100)]
```

Предв вывод: SU9 убыточен в принципе из-за малой пассажироемкости и высоких операционных расходов. Предложение: заменить на Боинг на рейсе в Белгород и сократить количество рейсов.
